```python
import os
import random
import pandas as pd
```

```python
stock_up=['BAC','TSLA','AMZN','GOOGL','GOOG','META','NVDA','T']

stock_sideway=['MU','C','GM','XOM','HAL','WMB','NEM','CVX']

stock_down=['F','KMI','MRO','COP','CTRA','AES','DVN','AXP']
```

```python
# for base command you have to follow the following template : "{absolute_path_to_python.exe} {absolute_path_to_your_working_directory}"

def test_model_with_random_parameters(gp_max_gen):
    for item in stock_up:
        os.system(f'python3 stgp.py {item} up {gp_max_gen}')
    for item in stock_sideway:
        os.system(f'python3 stgp.py {item} stock_sideway {gp_max_gen}')
    for item in stock_down:
        os.system(f'python3 stgp.py {item} down {gp_max_gen}')


def test_model_with_pso_parameters(const_multipleir, max_pso_itter):
    for item in stock_up:
        os.system(f'python3 stgp-cpso.py {item} up {const_multipleir} {max_pso_itter}')
    for item in stock_sideway:
        os.system(f'python3 stgp-cpso.py {item} stock_sideway {const_multipleir} {max_pso_itter}')
    for item in stock_down:
        os.system(f'python3 stgp-cpso.py {item} down {const_multipleir} {max_pso_itter}')
```

```python
#Start Random generated parameters(STGP)
test_model_with_random_parameters(20)
```

```python
#Start PSO generated parameters(STGP-CPSO)
pso_itters=[30,40,50]
for itter in pso_itters:
    test_model_with_pso_parameters(100,itter)
```
